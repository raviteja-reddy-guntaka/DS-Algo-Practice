```python
#TreeNode
class TreeNode:
    
    def __init__(self, name, title):
        self.name = name
        self.title = title
        self.children = []
        self.parent = None
        
    def add_child(self, child_node):
        self.children.append(child_node)
        child_node.parent = self
        
    def get_level(self):
        p = self.parent
        level = 0
        while p:
            level += 1
            p = p.parent
        return level
        
    def print_tree(self, mode):
        prefix = " " * 3 * self.get_level()
        prefix = prefix + "|__" if self.parent else ""
        suffix = ""
        if mode == "both":
            suffix = self.name + "(" + self.title + ")"
        elif mode == "name":
            suffix = self.name
        else:
            suffix = self.title
        print(prefix + suffix)
        if self.children:
            for ch in self.children:
                ch.print_tree(mode)
```

```python
def build_tree():
    root = TreeNode("Nilupul", "CEO")
    
    chinmay = TreeNode("Chinmay", "CTO")
    gels = TreeNode("Gels", "HR Head")
    
    root.add_child(chinmay)
    root.add_child(gels)
    
    vishwa = TreeNode("Vishwa", "Infrastructure Head")
    aamir = TreeNode("Aamir", "Application Head")
    chinmay.add_child(vishwa)
    chinmay.add_child(aamir)
    
    dhaval = TreeNode("Dhaval", "Cloud Manager")
    abhijit = TreeNode("Abhijit", "App Manager")
    vishwa.add_child(dhaval)
    vishwa.add_child(abhijit)
    
    peter = TreeNode("Peter", "Recruitment Manager")
    waqas = TreeNode("Waqas", "Policy Manager")
    gels.add_child(peter)
    gels.add_child(waqas)
    
    return root

if __name__=="__main__":
    root = build_tree()
    root.print_tree("name") # prints only name hierarchy
    print("")
    root.print_tree("designation") # prints only designation hierarchy
    print("")
    root.print_tree("both") # prints both (name and designation) hierarchy
```

```python
#TreeNode
class TreeNode:
    
    def __init__(self, name):
        self.name = name
        self.children = []
        self.parent = None
        
    def add_child(self, child_node):
        self.children.append(child_node)
        child_node.parent = self
        
    def get_level(self):
        p = self.parent
        level = 0
        while p:
            level += 1
            p = p.parent
        return level
        
    def print_tree(self, mode):
        if self.get_level() <= mode:
            prefix = " " * 3 * self.get_level()
            prefix = prefix + "|__" if self.parent else ""
            print(prefix + self.name)
            if self.children:
                for ch in self.children:
                    ch.print_tree(mode)
```
